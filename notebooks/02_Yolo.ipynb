```python
import os
import random
import shutil

BASE = '/content/drive/MyDrive/AI_BaiTapLon/cccd_extraction_src'

# TRAIN
train_image_dir = os.path.join(
    BASE,
    'data/detection/images/train'
)

train_label_dir = os.path.join(
    BASE,
    'data/detection/labels/train/Tu'
)

# VALIDATION
val_image_dir = os.path.join(
    BASE,
    'data/detection/images/val'
)

val_label_dir = os.path.join(
    BASE,
    'data/detection/labels/val'
)

os.makedirs(val_image_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)

# Lấy danh sách ảnh
images = [
    f for f in os.listdir(train_image_dir)
    if f.endswith(('.jpg', '.jpeg', '.png'))
]

random.shuffle(images)

# 20% validation
val_size = int(len(images) * 0.2)

val_images = images[:val_size]

print(f'Tổng ảnh: {len(images)}')
print(f'Số ảnh val: {len(val_images)}')

for img_name in val_images:

    label_name = os.path.splitext(img_name)[0] + '.txt'

    src_img = os.path.join(train_image_dir, img_name)
    src_label = os.path.join(train_label_dir, label_name)

    dst_img = os.path.join(val_image_dir, img_name)
    dst_label = os.path.join(val_label_dir, label_name)

    # Move image
    shutil.move(src_img, dst_img)

    # Move label
    if os.path.exists(src_label):
        shutil.move(src_label, dst_label)

print('Đã chia train/val')
```
