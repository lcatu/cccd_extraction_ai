```python
from google.colab import drive
drive.mount('/content/drive')

import os
import shutil
from tqdm import tqdm

# =========================
# BASE PATH
# =========================

BASE = '/content/drive/MyDrive/AI_BaiTapLon/cccd_extraction_src'

RAW_DIR = os.path.join(BASE, 'data/raw')

# =========================
# SOURCE FOLDERS
# =========================

FRONT_ROTATED_DIR = os.path.join(RAW_DIR, 'front_315_rotated')

CHIA_NGUOI_DIR = os.path.join(RAW_DIR, 'chia_nguoi')

# =========================
# OUTPUT FOLDER
# =========================

OUTPUT_DIR = os.path.join(RAW_DIR, 'data_gan_nhan')

# Tạo folder nếu chưa có
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f'Đã tạo folder:\n{OUTPUT_DIR}\n')

# =========================
# VALID IMAGE EXTENSIONS
# =========================

VALID_EXTENSIONS = ('.jpg', '.jpeg', '.png')

# =========================
# GET ALL IMAGE PATHS
# =========================

all_images = []

# Front rotated
for file_name in os.listdir(FRONT_ROTATED_DIR):

    if file_name.lower().endswith(VALID_EXTENSIONS):

        all_images.append(
            os.path.join(FRONT_ROTATED_DIR, file_name)
        )

# Chia nguoi/*
for person_folder in os.listdir(CHIA_NGUOI_DIR):

    person_path = os.path.join(CHIA_NGUOI_DIR, person_folder)

    if os.path.isdir(person_path):

        for file_name in os.listdir(person_path):

            if file_name.lower().endswith(VALID_EXTENSIONS):

                all_images.append(
                    os.path.join(person_path, file_name)
                )

# =========================
# SORT
# =========================

all_images = sorted(all_images)

print(f'Tổng số ảnh: {len(all_images)}\n')

# =========================
# COPY + RENAME
# =========================

copied = 0
failed = 0

for idx, src_path in enumerate(tqdm(all_images), start=1):

    try:

        ext = os.path.splitext(src_path)[1].lower()

        # Tên mới
        new_name = f'cccd_{idx:04d}{ext}'

        dst_path = os.path.join(OUTPUT_DIR, new_name)

        shutil.copy2(src_path, dst_path)

        copied += 1

    except Exception as e:

        print(f'Lỗi: {src_path}')
        print(e)

        failed += 1

# =========================
# DONE
# =========================

print('\n===== DONE =====')
print(f'Copied : {copied}')
print(f'Failed : {failed}')

print(f'\nSaved to:\n{OUTPUT_DIR}')
```

```python
from google.colab import drive
drive.mount('/content/drive')

import os
import shutil
from tqdm import tqdm

# =========================
# BASE PATH
# =========================

BASE = '/content/drive/MyDrive/AI_BaiTapLon/cccd_extraction_src'

RAW_DIR = os.path.join(BASE, 'data/raw')

# =========================
# SOURCE FOLDER
# =========================

SOURCE_DIR = os.path.join(RAW_DIR, 'data_gan_nhan')

# =========================
# OUTPUT FOLDER
# =========================

OUTPUT_DIR = os.path.join(RAW_DIR, 'chia_gan_nhan')

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# PEOPLE
# =========================

PEOPLE = [
    'Tu',
    'Chi',
    'Khoa',
    'TrangAnh'
]

# Tạo folder
for person in PEOPLE:

    os.makedirs(
        os.path.join(OUTPUT_DIR, person),
        exist_ok=True
    )

# =========================
# GET IMAGE LIST
# =========================

VALID_EXTENSIONS = ('.jpg', '.jpeg', '.png')

image_files = sorted([
    f for f in os.listdir(SOURCE_DIR)
    if f.lower().endswith(VALID_EXTENSIONS)
])

total_images = len(image_files)

print(f'Tổng số ảnh: {total_images}')

# =========================
# CALCULATE EVEN SPLIT
# =========================

base_count = total_images // len(PEOPLE)
remainder = total_images % len(PEOPLE)

print(f'Mỗi người tối thiểu: {base_count} ảnh')
print(f'Ảnh dư: {remainder}\n')

# =========================
# SPLIT
# =========================

start_idx = 0
copied = 0

for idx, person in enumerate(PEOPLE):

    # Người đầu tiên nhận thêm 1 ảnh nếu còn dư
    extra = 1 if idx < remainder else 0

    end_idx = start_idx + base_count + extra

    person_images = image_files[start_idx:end_idx]

    print(f'{person:<10}: {start_idx+1} → {end_idx} ({len(person_images)} ảnh)')

    for file_name in tqdm(person_images, desc=person):

        src_path = os.path.join(SOURCE_DIR, file_name)

        dst_path = os.path.join(
            OUTPUT_DIR,
            person,
            file_name
        )

        shutil.copy2(src_path, dst_path)

        copied += 1

    start_idx = end_idx

# =========================
# SUMMARY
# =========================

print('\n===== DONE =====')

for person in PEOPLE:

    person_dir = os.path.join(OUTPUT_DIR, person)

    count = len([
        f for f in os.listdir(person_dir)
        if f.lower().endswith(VALID_EXTENSIONS)
    ])

    print(f'{person:<10}: {count} ảnh')

print(f'\nTổng copied: {copied}')

print(f'\nSaved to:\n{OUTPUT_DIR}')
```
