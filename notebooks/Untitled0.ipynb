```python
import os
import random
import shutil
from google.colab import drive

# Ép mount lại để xóa toàn bộ cache cũ, đồng bộ tức thì với Drive Web
import os
from google.colab import drive

# Kiểm tra xem thư mục Drive của bạn thực sự đã xuất hiện chưa
if not os.path.exists('/content/drive/MyDrive'):
    print("Drive chưa được kết nối. Tiến hành mount...")
    drive.mount('/content/drive', force_remount=True)
else:
    print("✅ Drive thực ra đã kết nối sẵn và có dữ liệu từ trước rồi! Bỏ qua lệnh mount để tránh lỗi.")

# Tiếp tục chạy các dòng code cấu trúc bên dưới bình thường...
BASE = '/content/drive/MyDrive/AI_BaiTapLon/cccd_extraction_src'

# ==========================================
# ĐƯỜNG DẪN CẤU TRÚC CHUẨN (Khớp 100% với Drive của bạn)
# ==========================================

# 1. Thư mục GỐC chứa TOÀN BỘ ảnh CCCD bạn tải lên ban đầu (Thay đổi cho đúng nếu cần)
# Ví dụ hiện tại ảnh đang nằm hết trong thư mục 'train' của bạn:
source_image_dir = os.path.join(BASE, 'data/detection/images/train')

# 2. Thư mục nhãn GỐC tương ứng
source_label_dir = os.path.join(BASE, 'data/detection/labels/train')

# 3. Thư mục ĐÍCH muốn tách tập VAL sang (20%)
val_image_dir = os.path.join(BASE, 'data/detection/images/val')
val_label_dir = os.path.join(BASE, 'data/detection/labels/val')

# Tạo folder VAL nếu chưa có
os.makedirs(val_image_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)

# ==========================================
# QUÉT DỮ LIỆU TỪ FOLDER GỐC (source_image_dir)
# ==========================================
# Chuyển đuôi về chữ thường (.lower()) để tránh bỏ sót file .JPG hay .PNG viết hoa
images = [
    f for f in os.listdir(source_image_dir)
    if f.lower().endswith(('.jpg', '.jpeg', '.png'))
]

print(f"Tổng số ảnh gốc tìm thấy: {len(images)}")

# Trộn ngẫu nhiên danh sách ảnh
random.seed(42) # Giữ cố định kết quả trộn ở các lần chạy
random.shuffle(images)

# Lấy ra 20% số lượng ảnh để chuyển sang tập VAL
val_size = int(len(images) * 0.2)
val_imgs = images[:val_size]

print(f"Số lượng ảnh sẽ chuyển sang tập VAL (20%): {len(val_imgs)}")

# ==========================================
# TIẾN HÀNH DI CHUYỂN (MOVE) SANG VAL
# ==========================================
for img_name in val_imgs:
    # Đổi đuôi ảnh thành .txt để tìm file nhãn đi kèm
    label_name = os.path.splitext(img_name)[0] + '.txt'

    # Đường dẫn nguồn (đang ở trong folder train)
    src_img = os.path.join(source_image_dir, img_name)
    src_label = os.path.join(source_label_dir, label_name)

    # Đường dẫn đích (chuyển sang folder val)
    dst_img = os.path.join(val_image_dir, img_name)
    dst_label = os.path.join(val_label_dir, label_name)

    # Di chuyển file ảnh (Dùng shutil.move để chuyển hẳn đi, không dùng copy)
    if os.path.exists(src_img):
        shutil.move(src_img, dst_img)

    # Di chuyển file nhãn (nếu có)
    if os.path.exists(src_label):
        shutil.move(src_label, dst_label)

print("--- Chia tập train/val bằng phương pháp tách từ folder gốc hoàn tất! ---")
```
