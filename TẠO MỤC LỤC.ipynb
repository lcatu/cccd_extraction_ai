```python
import os
from google.colab import drive
drive.mount('/content/drive')

BASE = '/content/drive/MyDrive/AI_BaiTapLon/cccd_extraction_src'

def print_tree(base_path, max_files=3):
    """
    In cấu trúc thư mục dạng cây.
    max_files: số file tối đa hiển thị trong mỗi folder (tránh spam)
    """
    base_path = os.path.abspath(base_path)
    basename  = os.path.basename(base_path)
    print(f'{basename}/')

    def _walk(path, prefix=''):
        entries = sorted(os.listdir(path))
        dirs    = [e for e in entries if os.path.isdir(os.path.join(path, e))]
        files   = [e for e in entries if os.path.isfile(os.path.join(path, e))]

        # In folders
        for i, d in enumerate(dirs):
            is_last_dir  = (i == len(dirs) - 1) and len(files) == 0
            connector    = '└── ' if is_last_dir else '├── '
            child_prefix = '    ' if is_last_dir else '│   '
            sub_path     = os.path.join(path, d)
            count        = len(os.listdir(sub_path))
            suffix       = f'  ({count} items)' if count > 0 else '  (trống)'
            print(f'{prefix}{connector}{d}/{suffix}')
            _walk(sub_path, prefix + child_prefix)

        # In files (giới hạn max_files)
        for i, f in enumerate(files[:max_files]):
            is_last = (i == min(max_files, len(files)) - 1)
            connector = '└── ' if is_last and len(files) <= max_files else '├── '
            size = os.path.getsize(os.path.join(path, f))
            size_str = (f'{size/1024/1024:.1f} MB' if size > 1024*1024
                        else f'{size/1024:.1f} KB' if size > 1024
                        else f'{size} B')
            print(f'{prefix}{connector}{f}  [{size_str}]')

        # Nếu có nhiều file hơn max_files → thông báo
        if len(files) > max_files:
            print(f'{prefix}└── ... và {len(files) - max_files} file khác')

    _walk(base_path)
    print()
    print(f'Tổng: {sum(len(f) for _, _, f in os.walk(base_path))} files, '
          f'{sum(len(d) for _, d, _ in os.walk(base_path))} folders')

# Chạy
print_tree(BASE, max_files=3)
```
