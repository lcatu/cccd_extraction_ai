```python
from PIL import Image
import cv2
import numpy as np

IMG_PATH = '/content/drive/MyDrive/AI_BaiTapLon/cccd_extraction_src/data/raw/chia_gan_nhan/TrangAnh/cccd_0864.jpg'

results = model.predict(IMG_PATH, conf=0.5)
result = results[0]

image = cv2.imread(IMG_PATH)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Lấy tên class
class_names = result.names  # {0: 'card', 1: 'dob', ...}

# =========================
# CẮT TỪNG FIELD
# =========================

fields = {}

for box in result.boxes:
    cls_id = int(box.cls)
    cls_name = class_names[cls_id]
    conf = float(box.conf)

    # Bỏ qua class 'card' (bbox toàn thẻ)
    if cls_name == 'card':
        continue

    x1, y1, x2, y2 = map(int, box.xyxy[0])

    # Thêm padding nhỏ cho OCR dễ đọc hơn
    PAD = 4
    x1 = max(0, x1 - PAD)
    y1 = max(0, y1 - PAD)
    x2 = min(image.shape[1], x2 + PAD)
    y2 = min(image.shape[0], y2 + PAD)

    crop = image_rgb[y1:y2, x1:x2]

    # Nếu 1 class detect 2 lần (residence), gộp theo conf cao hơn
    if cls_name not in fields or conf > fields[cls_name]['conf']:
        fields[cls_name] = {
            'crop': crop,
            'conf': conf,
            'bbox': (x1, y1, x2, y2)
        }

# =========================
# OCR TỪNG FIELD
# =========================

import easyocr
reader = easyocr.Reader(['vi', 'en'], gpu=True)

ocr_results = {}

for field_name, field_data in fields.items():
    crop = field_data['crop']

    # EasyOCR nhận numpy array trực tiếp
    texts = reader.readtext(crop, detail=0, paragraph=True)
    text = ' '.join(texts).strip()

    ocr_results[field_name] = text
    print(f"{field_name:15s}: {text}")

print("\n--- KẾT QUẢ ---")
for k, v in ocr_results.items():
    print(f"{k}: {v}")
```
